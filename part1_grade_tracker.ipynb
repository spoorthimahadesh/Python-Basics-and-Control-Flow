```python
#TASK !
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
#Q1
cleaned_students = []

for student in raw_students:
     
    cleaned = {}  
    name = student["name"]
    name = name.strip()       
    name = name.title()       
    cleaned["name"] = name
     # convert roll to int
    roll = student["roll"]
    roll = int(roll)          
    cleaned["roll"] = roll
     # convert marks string to list of ints
    marks_str = student["marks_str"]
    marks_split = marks_str.split(",")   
     
    marks = []
    for m in marks_split:
        marks.append(int(m))  
    
    cleaned["marks"] = marks
    
    cleaned_students.append(cleaned)
# print cleaned data
for cleaned_list in cleaned_students:
    print(cleaned_list)
```

```python
#Q3
new_count = 0

while True:
    subject = input("Subject name (or type 'done' to stop): ")
    if subject.lower() == "done":
        break
    mark = input("Enter marks (0-100): ")
    
    if not mark.isdigit():
        print("Invalid marks! Enter numeric marks.")
        continue
        
    x = int(mark)
    if x < 0 or x > 100:
        print("Marks should be between 0 and 100.")
        continue

    subjects.append(subject)
    marks.append(x)
    new_count += 1

total = sum(marks)
average = total / len(marks)

print("\nTotal new subjects added:", new_count)
print("Updated average score:", average)
```
